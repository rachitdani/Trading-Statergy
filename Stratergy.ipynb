```python
import requests
import pandas as pd

class ScriptData:
    def __init__(self,api_key):
        self.api_key = api_key
        self.base_url = 'https://www.alphavantage.co/query'

    def fetch_intraday_data(self, script):
        params = {
            'function': 'TIME_SERIES_INTRADAY',
            'symbol': script,
            'interval': '1min',
            'apikey': 'GTXEAPKTO18LALTS'
        }
        response = requests.get(self.base_url, params=params)
        data = response.json()
        print(data)#for getting the appropriate error message as the api allows only 5 calls per minute and also printing the metadata [We can comment out if not needed]
        return data['Time Series (1min)']

    def convert_intraday_data(self, script):
        intraday_data = self.fetch_intraday_data(script)
        df = pd.DataFrame.from_dict(intraday_data, orient='index')
        df.columns = ['open', 'high', 'low', 'close', 'volume']
        df.index = pd.to_datetime(df.index)
        df = df.astype({'open': float, 'high': float, 'low': float, 'close': float, 'volume': int})
        df['timestamp'] = df.index
        return df

    def __getitem__(self, key):
        return self.convert_intraday_data(key)

    def __setitem__(self, key, value):
        raise NotImplementedError("Setting item is not supported.")

    def __contains__(self, key):
        try:
            self.convert_intraday_data(key)
            return True
        except KeyError:
            return False
```

```python
def indicator1(df, timeperiod):
    indicator_df = pd.DataFrame()
    indicator_df['timestamp'] = df['timestamp']
    indicator_df['indicator'] = df['close'].rolling(timeperiod).mean()
    return indicator_df
```

```python
class Strategy:
    def __init__(self):
        self.script_data = ScriptData('GTXEAPKTO18LALTS')

    def generate_signals(self, script):
        df = self.script_data.convert_intraday_data(script)
        close_data = df['close']
        indicator_data = indicator1(df, timeperiod=5)['indicator']

        signals = pd.DataFrame()
        signals['timestamp'] = df.index
        signals['signal'] = 'NO_SIGNAL'

        # Compute signals based on indicator and close data
        for i in range(1, len(df)):
            if indicator_data[i] > close_data[i] and indicator_data[i-1] <= close_data[i-1]:
                signals.loc[i, 'signal'] = 'BUY'
            elif indicator_data[i] < close_data[i] and indicator_data[i-1] >= close_data[i-1]:
                signals.loc[i, 'signal'] = 'SELL'

        return signals

    def print_buy_sell_signals(self, script):
        signals = self.generate_signals(script)
        buy_sell_signals = signals[signals['signal'].isin(['BUY', 'SELL'])]
        print(buy_sell_signals)
```
